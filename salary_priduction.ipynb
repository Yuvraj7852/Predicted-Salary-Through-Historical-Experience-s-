Predicted Salay Through Historical Experience's

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("salary_priduction.csv")
```

```python
df.head()
```

```python
#select rows
x=df.iloc[:,0].values
y=df.iloc[:,1].values
print(x)
print(y)
```

```python
x=x.reshape(-1,1)
print(x)
```

```python
#split data for testing and training purpose
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
plt.scatter(x_train,y_train,color='r', marker='o')
plt.xlabel('expreince')
plt.ylabel('price')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#reg = LinearRegression()
#reg = DecisionTreeRegressor()
#reg =RandomForestRegressor()
#from sklearn.linear_model import dec
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_predict=reg.predict(x_test)
```

```python
plt.scatter(x_train,y_train,color='g', marker='o')
plt.xlabel('expreince')
plt.ylabel('price')
plt.plot(x_train,reg.predict(x_train))
plt.show()
```
